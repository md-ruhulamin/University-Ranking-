```python
import pandas as pd

df = pd.read_csv('/content/World University Rankings.csv')
```

```python
df.head(20)
```

```python
import matplotlib.pyplot as plt

top_unis = df[['name', 'scores_teaching', 'scores_research', 'scores_citations']].set_index('name')
top_unis.plot(kind='bar', figsize=(12, 6))
plt.title("Teaching, Research, and Citation Scores by University")
plt.ylabel("Score")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
labels = ['Female', 'Male']
ratios = [49, 51]

plt.pie(ratios, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Female to Male Ratio - University of Oxford')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='scores_industry_income', y='scores_international_outlook', hue='location')
plt.title("International Outlook vs. Industry Income")
plt.xlabel("Industry Income Score")
plt.ylabel("International Outlook Score")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df_sorted = df.sort_values('stats_number_students', ascending=True)
plt.barh(df_sorted['name'], df_sorted['stats_number_students'])
plt.xlabel("Number of Students")
plt.title("Student Population per University")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

score_columns = ['scores_teaching', 'scores_research', 'scores_citations', 'scores_industry_income', 'scores_international_outlook']
correlation = df[score_columns].corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Scores")
plt.show()
```
